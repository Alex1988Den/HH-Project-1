```python
1. Начнем с простого - с признака **"Образование и ВУЗ"**. Его текущий формат это: **<Уровень образования год выпуска ВУЗ специальность...>**. Например:
* Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)...
* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

*Совет: проверяйте полученные категории, например, с помощью метода unique()*
```

```python
import pandas as pd

# Шаг 1: Загрузим данные (предположим, что у нас есть DataFrame с названием df и столбцом "Образование и ВУЗ")
data = pd.read_csv("C:/Users/user/OneDrive/Desktop/jupyter/hh_database.csv", sep=";")

# Пример данных (закомментируйте предыдущую строку и раскомментируйте следующую, если у вас нет реальных данных)
data = {'Образование и ВУЗ': ['Высшее образование 2016 Московский авиационный институт (национальный исследовательский университет)',
                             'Неоконченное высшее образование 2000 Балтийская государственная академия рыбопромыслового флота']}
df = pd.DataFrame(data)

# Шаг 2: Создаем функцию-преобразование для определения уровня образования
def extract_education_level(text):
    words = text.split()
    if "Высшее" in words:
        return "высшее"
    elif "Неоконченное" in words:
        return "неоконченное высшее"
    elif "специальное" in words:
        return "среднее специальное"
    else:
        return "среднее"

# Шаг 3: Применяем функцию-преобразование к столбцу "Образование и ВУЗ" и создаем новый столбец "Образование"
df['Образование'] = df['Образование и ВУЗ'].apply(extract_education_level)

# Шаг 4: Удаляем столбец "Образование и ВУЗ"
df.drop('Образование и ВУЗ', axis=1, inplace=True)

# Шаг 5: Проверяем уникальные значения нового столбца "Образование"
print(df['Образование'].unique())
```

```python
2. Теперь нас интересует столбец **"Пол, возраст"**. Сейчас он представлен в формате **<Пол , возраст , дата рождения >**. Например:
* Мужчина , 39 лет , родился 27 ноября 1979 
* Женщина , 21 год , родилась 13 января 2000
Как вы понимаете, нам необходимо выделить каждый параметр в отдельный столбец.

Создайте два новых признака **"Пол"** и **"Возраст"**. При этом важно учесть:
* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина. 
* Признак возраста должен быть представлен целыми числами.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

*Совет: обратите внимание на структуру текста в столбце, в части на то, как разделены параметры пола, возраста и даты рождения между собой - символом ' , '. 
Гарантируется, что структура одинакова для всех строк в таблице. Вы можете воспользоваться этим.*
```

```python
import pandas as pd

# Шаг 1: Загрузим данные (предположим, что у нас есть DataFrame с названием df и столбцом "Пол, возраст")
data = pd.read_csv("C:/Users/user/OneDrive/Desktop/jupyter/hh_database.csv", sep=";")

data = {'Пол, возраст': ['Мужчина , 39 лет , родился 27 ноября 1979',
                        'Женщина , 21 год , родилась 13 января 2000']}
df = pd.DataFrame(data)

# Шаг 2: Создаем функцию-преобразование для извлечения значения пола и возраста
def extract_gender_age(text):
    gender, rest = text.split(' , ', 1)  # Ограничим разделение только первым символом ' , '
    gender = 'М' if 'Мужчина' in gender else 'Ж'
    
    # Используем регулярное выражение для поиска чисел в строке возраста
    import re
    age = int(re.search(r'\d+', rest).group())
    
    return gender, age

# Шаг 3: Применяем функцию-преобразование к столбцу "Пол, возраст" и создаем два новых столбца "Пол" и "Возраст"
df['Пол'], df['Возраст'] = zip(*df['Пол, возраст'].apply(extract_gender_age))

# Шаг 4: Удаляем столбец "Пол, возраст"
df.drop('Пол, возраст', axis=1, inplace=True)

# Выведем результат
print(df)
```

```python
3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.
```

```python
import pandas as pd
import numpy as np
import re

# Шаг 1: Загрузим данные 
data = pd.read_csv("C:/Users/user/OneDrive/Desktop/jupyter/hh_database.csv", sep=";")
df = pd.DataFrame(data)

# Шаг 2: Создаем функцию-преобразование для извлечения информации об опыте работы в месяцах
def extract_experience_in_months(text):
    if pd.isna(text) or "Не указано" in text:
        return np.nan

    total_months = 0
    years_match = re.search(r"(\d+)\s*год", text)
    if years_match:
        years = int(years_match.group(1))
        total_months += years * 12

    months_match = re.search(r"(\d+)\s*месяц", text)
    if months_match:
        months = int(months_match.group(1))
        total_months += months

    return total_months

# Шаг 3: Применяем функцию-преобразование к столбцу "Опыт работы" и создаем новый столбец "Опыт работы (месяц)"
df['Опыт работы (месяц)'] = df['Опыт работы'].apply(extract_experience_in_months)

# Шаг 4: Удаляем столбец "Опыт работы"
df.drop('Опыт работы', axis=1, inplace=True)

# Выведем результат
print(df)
```

```python
4. Хорошо идем! Следующий на очереди признак "Город, переезд, командировки". Информация в нем представлена в следующем виде: **<Город , (метро) , готовность к переезду (города для переезда) , готовность к командировкам>**. В скобках указаны необязательные параметры строки. Например, можно встретить следующие варианты:

* Москва , не готов к переезду , готов к командировкам
* Москва , м. Беломорская , не готов к переезду, не готов к командировкам
* Воронеж , готов к переезду (Сочи, Москва, Санкт-Петербург) , готов к командировкам

Создадим отдельные признаки **"Город"**, **"Готовность к переезду"**, **"Готовность к командировкам"**. При этом важно учесть:

* Признак **"Город"** должен содержать только 4 категории: "Москва", "Санкт-Петербург" и "город-миллионник" (их список ниже), остальные обозначьте как "другие".

    Список городов-миллионников:
    
   <code>million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
    </code>
    Инфорация о метро, рядом с которым проживает соискатель нас не интересует.
* Признак **"Готовность к переезду"** должен иметь два возможных варианта: True или False. Обратите внимание, что возможны несколько вариантов описания готовности к переезду в признаке "Город, переезд, командировки". Например:
    * … , готов к переезду , …
    * … , не готова к переезду , …
    * … , готова к переезду (Москва, Санкт-Петербург, Ростов-на-Дону)
    * … , хочу переехать (США) , …
    
    Нас интересует только сам факт возможности или желания переезда.
* Признак **"Готовность к командировкам"** должен иметь два возможных варианта: True или False. Обратите внимание, что возможны несколько вариантов описания готовности к командировкам в признаке "Город, переезд, командировки". Например:
    * … , готов к командировкам , … 
    * … , готова к редким командировкам , …
    * … , не готов к командировкам , …
    
    Нас интересует только сам факт готовности к командировке.
    
    Еще один важный факт: при выгрузки данных у некоторых соискателей "потерялась" информация о готовности к командировкам. Давайте по умолчанию будем считать, что такие соискатели не готовы к командировкам.
    
Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*
```

```python
import pandas as pd
import numpy as np
import re

# Шаг 1: Загрузим данные (предположим, что у нас есть DataFrame с названием df и столбцом "Город, переезд, командировки")
data = pd.read_csv("C:/Users/user/OneDrive/Desktop/jupyter/hh_database.csv", sep=";")

df = pd.DataFrame(data)

# Шаг 2: Создаем функции-преобразования для извлечения информации о городе, готовности к переезду и готовности к командировкам
def extract_city(text):
    # Используем регулярное выражение для поиска города
    city_match = re.search(r'[А-Яа-я\-]+', text)
    if city_match:
        city = city_match.group()
        # Проверяем, является ли город миллионником
        million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара',
                          'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград']
        if city in million_cities:
            return 'город-миллионник'
        elif city == 'Москва' or city == 'Санкт-Петербург':
            return city
        else:
            return 'другой'
    return np.nan

def extract_relocation(text):
    if "готов" in text.lower() and "не готов" not in text.lower():
        return True
    elif "хочу переехать" in text.lower():
        return True
    return False

def extract_business_trip(text):
    if "готов" in text.lower():
        return True
    return False

# Шаг 3: Применяем функции-преобразования к столбцу "Город, переезд, командировки" и создаем новые столбцы "Город", "Готовность к переезду" и "Готовность к командировкам"
df['Город'] = df['Город, переезд, командировки'].apply(extract_city)
df['Готовность к переезду'] = df['Город, переезд, командировки'].apply(extract_relocation)
df['Готовность к командировкам'] = df['Город, переезд, командировки'].apply(extract_business_trip)

# Шаг 4: Удаляем столбец "Город, переезд, командировки"
df.drop('Город, переезд, командировки', axis=1, inplace=True)

# Выведем результат
print(df)
```

```python
5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы
```

```python
Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".
```

```python
import pandas as pd
import re

# Шаг 1: Загрузим данные с курсами валют
currency_data = pd.read_csv("C:/Users/user/OneDrive/Desktop/jupyter/exchangerates.csv")

# Шаг 2: Загрузим данные из файла hh_database.csv
df = pd.read_csv("C:/Users/user/OneDrive/Desktop/jupyter/hh_database.csv", sep=";")

# Шаг 3: Преобразуем столбец "Обновление резюме" в тип данных datetime и достанем дату
df['Обновление резюме'] = pd.to_datetime(df['Обновление резюме'])
currency_data['date'] = pd.to_datetime(currency_data['date'])

# Шаг 4: Преобразовать столбец "ЗП" обратно в строковый формат и избавиться от некорректных значений
df['ЗП'] = df['ЗП'].astype(str).str.replace('[^\d]', '', regex=True)

# Выделить сумму желаемой заработной платы и наименование валюты
df['ЗП'] = df['ЗП'].astype(float)
df['Валюта'] = df['ЗП'].astype(str).str.extract(r'([а-яa-z]+)', flags=re.IGNORECASE)

# Маппинг наименования валюты в стандарт ISO согласно таблице
currency_mapping = {
    'руб.': 'RUB',
    'руб': 'RUB',
    'грн': 'UAH',
    'eur': 'EUR',
    'usd': 'USD',
    # Добавьте остальные валюты согласно таблице
}
df['Валюта'] = df['Валюта'].str.lower().map(currency_mapping)

# Convert the 'Валюта' column to the same type as the 'currency_data['currency']' column
df['Валюта'] = df['Валюта'].astype(object)

# Шаг 5: Присоединить таблицу с курсами по дате и названию валюты
df_merged = pd.merge(df, currency_data, left_on=['Обновление резюме', 'Валюта'], right_on=['date', 'currency'], how='left')

# Шаг 6: Заполнить значение close для рубля единицей (1) и рассчитать новый столбец "ЗП (руб)"
df_merged['close'] = df_merged['close'].fillna(1)
df_merged['ЗП (руб)'] = df_merged['ЗП'] * df_merged['close'] / df_merged['proportion']

# Шаг 7: Удалить исходный столбец "ЗП" и все промежуточные столбцы
df_merged.drop(['ЗП', 'date', 'proportion', 'close', 'Валюта', 'currency'], axis=1, inplace=True)

# Выведем результат
print(df_merged)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных из CSV-файла с разделителем ";"
df = pd.read_csv("C:/Users/user/OneDrive/Desktop/jupyter/hh_database.csv", sep=";", engine='python')

# Преобразование столбца "ЗП" обратно в строковый формат и избавление от некорректных символов
df['ЗП'] = df['ЗП'].astype(str).str.replace('[^\d]', '', regex=True)

# Преобразование столбца "ЗП" в числовой формат (int)
df['ЗП'] = df['ЗП'].astype(int)

# Фильтрация данных по желаемой заработной плате (ЗП) больше 0
df_filtered = df[df['ЗП'] > 0]

# Преобразование столбца "Образование и ВУЗ" для получения отдельных столбцов "Образование" и "ВУЗ"
df_filtered.loc[:, ['Образование', 'ВУЗ']] = df_filtered['Образование и ВУЗ'].str.split(',', n=1, expand=True)

# Удаление столбца "Образование и ВУЗ", так как мы получили отдельные столбцы "Образование" и "ВУЗ"
df_filtered.drop(columns=['Образование и ВУЗ'], inplace=True)

# Группировка данных по столбцу "Образование" и нахождение медианного значения ЗП
median_salary_by_education = df_filtered.groupby('Образование')['ЗП'].median()

# Построение столбчатой диаграммы
plt.bar(median_salary_by_education.index, median_salary_by_education.values, color="skyblue")

# Настройка названий и подписей на оси x и y
plt.xticks(rotation=90)
plt.xlabel("Уровень образования")
plt.ylabel("Медианная ЗП")
plt.title("Зависимость медианной ЗП от уровня образования")

# Отображение диаграммы
plt.show()
```

```python
На основе представленной диаграммы "Зависимость медианной ЗП от уровня образования" мы можем сделать следующие выводы:

    Наибольшие уровни желаемой заработной платы наблюдаются для следующих уровней образования (по убыванию медианной ЗП):
        Высшее образование (специалист, бакалавр, магистр)
        Незаконченное высшее образование
        Второе высшее образование

    Наименьшие уровни желаемой заработной платы наблюдаются для следующих уровней образования (по возрастанию медианной ЗП):
        Начальное образование
        Среднее образование
        Нет образования

Таким образом, уровень образования явно влияет на уровень желаемой заработной платы. Люди с высшим образованием,
в целом, ожидают более высокую заработную плату, чем люди с низшим образованием или без образования.
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Параметры для чтения файла CSV по частям
file_path = "C:/Users/user/OneDrive/Desktop/jupyter/hh_database.csv"
chunk_size = 10000

# Чтение файла CSV по частям
chunks = pd.read_csv(file_path, sep=";", chunksize=chunk_size)

# Список для хранения сводных таблиц чанков
pivot_tables = []

# Обработка данных по частям
for chunk in chunks:
    # Шаг 1: Проверим, что данные были успешно загружены
    print(chunk.head())

    # Шаг 2: Проверим информацию о DataFrame, чтобы убедиться, что все столбцы прочитаны правильно
    print(chunk.info())

    # Шаг 3: Проверим, что столбец 'Образование' существует в текущем чанке
    if 'Образование' in chunk.columns:
        # Создадим сводную таблицу с медианной желаемой заработной платой, сгруппированной по возрасту и образованию
        pivot_table = chunk.pivot_table(values='ЗП', index='Пол, возраст', columns='Образование', aggfunc='median')
        pivot_tables.append(pivot_table)
    else:
        print("Warning: 'Образование' column not found in the current chunk.")

# Проверим, что хотя бы одна сводная таблица была создана
if pivot_tables:
    # Объединим сводные таблицы чанков в одну итоговую сводную таблицу
    final_pivot_table = pd.concat(pivot_tables)

    # Шаг 4: Построим тепловую карту с использованием библиотеки seaborn
    plt.figure(figsize=(10, 8))
    sns.heatmap(final_pivot_table, annot=True, cmap='YlGnBu', fmt='.0f', cbar=True)
    plt.title('Зависимость медианной заработной платы от пола, возраста и образования')
    plt.xlabel('Образование')
    plt.ylabel('Пол, возраст')
    plt.show()
else:
    print("No pivot tables were created due to missing 'Образование' column.")
```

```python
1.Диаграмма рассеяния (scatter plot) для сравнения возраста и заработной платы.
2.График распределения заработной платы по полу.
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загружаем данные из файла в DataFrame
file_path = "C:/Users/user/OneDrive/Desktop/jupyter/hh_database.csv"
df = pd.read_csv(file_path, sep=';')

# Удаляем строки, где столбец 'ЗП' имеет тип данных str и не содержит ' руб.'
df = df[df['ЗП'].apply(lambda x: isinstance(x, str) and ' руб.' in x)]

# Удаляем подстроку ' руб.' из столбца 'ЗП' и преобразуем его в числовой тип
df['ЗП'] = df['ЗП'].apply(lambda x: float(x.replace(' руб.', '')))

# Удаляем лишние пробелы из названия столбца 'Пол, возраст'
df.rename(columns={'Пол, возраст': 'Пол,возраст'}, inplace=True)

# Разделяем столбец "Пол, возраст" на отдельные столбцы "Пол" и "Возраст"
split_data = df['Пол,возраст'].str.split(', ', expand=True)

# Remove leading and trailing spaces from the 'Пол' and 'Возраст' columns
split_data[0] = split_data[0].str.strip()
split_data[1] = split_data[1].str.strip()

# Assign 'Пол' and 'Возраст' columns to the original DataFrame
df['Пол'] = split_data[0]
df['Возраст'] = split_data[1]

# Убираем лишние символы из столбца "Возраст" и преобразуем его в числовой тип
df['Возраст'] = df['Возраст'].str.extract(r'(\d+)').astype(int)

# Создадим возрастовые категории
bins = [0, 18, 25, 35, 50, 100]
labels = ['до 18', '18-25', '25-35', '35-50', '50+']
df['Возрастовая категория'] = pd.cut(df['Возраст'], bins=bins, labels=labels)

# 1. Диаграмма рассеяния для сравнения возраста и заработной платы
plt.figure(figsize=(10, 6))
plt.scatter(df['Возраст'], df['ЗП'], alpha=0.5, color='blue')
plt.xlabel('Возраст')
plt.ylabel('Заработная плата')
plt.title('Сравнение возраста и заработной платы')
plt.grid(True)
plt.show()

# 2. График распределения заработной платы по полу
plt.figure(figsize=(8, 6))
df.groupby('Пол')['ЗП'].plot(kind='hist', alpha=0.5, legend=True)
plt.xlabel('Заработная плата')
plt.ylabel('Количество')
plt.title('Распределение заработной платы по полу')
plt.legend(title='Пол')
plt.grid(True)
plt.show()
```

```python
import pandas as pd

# Загрузка данных из CSV-файла с разделителем ";"
data = pd.read_csv("C:/Users/user/OneDrive/Desktop/jupyter/hh_database.csv", sep=";")

# Определение правильного имени столбцов для опыта работы и возраста и удаление начальных/конечных пробелов, если они есть
experience_column_name = 'Опыт работы'
age_column_name = 'Пол, возраст'
data[experience_column_name] = data[experience_column_name].str.strip()
data[age_column_name] = data[age_column_name].str.strip()

# Замена пропущенных значений на пустые строки
data[experience_column_name].fillna('', inplace=True)

# Функция для извлечения числового значения опыта работы в месяцах
def extract_experience_months(experience_str):
    words = experience_str.split()
    if 'лет' in words and 'месяц' in words:
        years_index = words.index('лет') - 1
        months_index = words.index('месяц') - 1
        if years_index >= 0 and months_index >= 0 and words[years_index].replace(',', '').isdigit() and words[months_index].isdigit():
            years = int(words[years_index].replace(',', ''))
            months = int(words[months_index])
            return years * 12 + months
    elif 'лет' in words and words.index('лет') - 1 >= 0 and words[words.index('лет') - 1].replace(',', '').isdigit():
        years = int(words[words.index('лет') - 1].replace(',', ''))
        return years * 12
    elif 'месяц' in words and words.index('месяц') - 1 >= 0 and words[words.index('месяц') - 1].isdigit():
        months = int(words[words.index('месяц') - 1])
        return months
    return 0

# Функция для извлечения числового значения возраста
def extract_age(age_str):
    age_parts = age_str.split()
    if len(age_parts) > 1 and age_parts[1].replace(',', '').isdigit():
        return int(age_parts[1].replace(',', ''))
    return 0

# Применение функций для извлечения числовых значений опыта работы и возраста
data['Опыт работы (месяц)'] = data[experience_column_name].apply(extract_experience_months)
data['Возраст (лет)'] = data[age_column_name].apply(extract_age)

# Нахождение резюме, в которых опыт работы в годах превышает возраст соискателя
outliers = data[data['Опыт работы (месяц)'] // 12 > data['Возраст (лет)']]

# Удаление резюме, в которых опыт работы в годах превышает возраст соискателя
data = data.drop(outliers.index)

# Вывод обновленного количества строк в данных после удаления резюме
print(data.shape[0])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re  # Добавляем модуль re для работы с регулярными выражениями

# Загрузка данных из CSV-файла
data = pd.read_csv("C:/Users/user/OneDrive/Desktop/jupyter/hh_database.csv", sep=";")

# Определение правильного имени столбца для возраста и удаление начальных/конечных пробелов, если они есть
age_column_name = 'Пол, возраст'
data[age_column_name] = data[age_column_name].str.strip()

# Извлечение информации о возрасте из столбца "Пол, возраст" и преобразование в числовой тип данных
def extract_age(age_str):
    age_match = re.search(r'(\d+)', age_str)  # Ищем числовое значение возраста в строке
    if age_match:
        return int(age_match.group())  # Возвращаем найденное числовое значение возраста
    return None

data['Возраст'] = data[age_column_name].apply(extract_age)

# Удаление строк с отсутствующими или недопустимыми значениями возраста
data = data.dropna(subset=['Возраст'])
data = data[data['Возраст'] > 0]  # Удаление строк с нулевым или отрицательным возрастом

# Преобразование столбца возраста в логарифмический масштаб
fig, ax = plt.subplots(1, 1, figsize=(8, 4))
log_age = np.log(data['Возраст'] + 1)
histplot = sns.histplot(log_age, bins=30, ax=ax)

# Добавление линий для среднего значения и границ интервала метода трех сигм
histplot.axvline(log_age.mean(), color='k', lw=2, label='Среднее значение')
histplot.axvline(log_age.mean() + 4 * log_age.std(), color='r', linestyle='--', lw=2, label='3-Sigma Верхняя граница')
histplot.axvline(log_age.mean() - 3 * log_age.std(), color='r', linestyle='--', lw=2, label='3-Sigma Нижняя граница')
histplot.set_title('Распределение возраста в логарифмическом масштабе')
histplot.set_xlabel('Логарифм Возраста')
histplot.set_ylabel('Плотность')
histplot.legend()

plt.show()

# Функция для определения выбросов по методу z-отклонения
def outliers_z_score_mod(data, feature, left=3, right=3, log_scale=False):
    if log_scale:
        x = np.log(data[feature] + 1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned

# Идентификация выбросов в возрасте с использованием метода z-отклонения (правый хвост)
outliers, cleaned_data = outliers_z_score_mod(data, 'Возраст', left=3, right=4, log_scale=True)

# Вывод таблицы с выбросами
print(outliers)
```
